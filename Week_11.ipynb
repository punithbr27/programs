Impoting a file


````python
from nltk import sent_tokenize

file = open("C:\\Users\\Admin\\Desktop\\a.txt",'r') 
text = file.read() 
text
````

Sentence tokenizer

````python
sent_token = sent_tokenize(text)
sent_token

print("Number of sentence are ",len(sent_token))
print()

for i in range(len(sent_token)):
    print("Sentence ",i,' : ',sent_token[i])
    print("\n")
````

Word Tokenize

````python
from nltk.tokenize import word_tokenize

word_token = word_tokenize(text,language='english',preserve_line=False)
print("Number of words in the paragraph is ",len(word_token))
print(word_token)
````

N-Grams


````python
from nltk.util import ngrams

print("Bi - Grams : \n\n",list(ngrams(word_token,2)))
print("Tri - Grams : \n\n",list(ngrams(word_token,3)))
````

Visualizing Frequency distribution of words

````python
from nltk.probability import FreqDist

file = open("C:\\Users\\Admin\\Desktop\\a.txt",'r')
text = file.read()

all_fdist = FreqDist(word_token)
all_fdist

import matplotlib.pyplot as plt
import pandas as pd

all_fdist = pd.Series(dict(all_fdist))
print(all_fdist)

plt.figure(figsize=(25,10))
all_fdist.plot(kind = 'bar')
plt.title("Frequency Distribution of Words")
plt.xlabel("Words")
plt.ylabel("Count of the words")
plt.tight_layout()
plt.show()
````

Removing special characters, numbers in string and lowering them ```"Stemming and Lemmatization"```


````python
import re

text = re.sub('[^A-Za-z0-9]+',' ',text)
print(text)
text = text.lower()
print("\n",text)
text = re.sub("\S*\d\S"," ",text).strip()
text
````

Visualizing after cleaning


````python
import nltk
word_token = word_tokenize(text)
stop_words = nltk.corpus.stopwords.words('english')
words_sw_removed = []
# stop_words_present = []
for word in word_token:
    if word in stop_words:
        # stop_words_present.append(word)
        pass
    else:
        words_sw_removed.append(word)
print(words_sw_removed)
# print(stop_words_present)
# print(stop_words)

all_fdist = FreqDist(words_sw_removed).most_common(200)
all_fdist = pd.Series(dict(all_fdist))

plt.figure(figsize=(15,10))
all_fdist.plot(kind='bar')
plt.show()
# all_fdist
````

Visualize with word cloud


````python
from wordcloud import WordCloud, STOPWORDS

stop_words = set(STOPWORDS)

wordcloud = WordCloud(width=800,height=800,
                      background_color='black',
                      stopwords=stop_words,
                      min_font_size=10).generate(text)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
````
