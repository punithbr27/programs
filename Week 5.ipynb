    Extract the upper and lower quantiles

```python
dist_lq = data['distance_from_home'].quantile(.25)
dist_uq = data['distance_from_home'].quantile(.75)
# print("Quartile ranges till \n",data['distance_from_home'].quantile(.75),"\n\n")
 
dist_IQR = dist_uq - dist_lq

lower_bound = dist_lq - 1.5 * dist_IQR
upper_bound = dist_uq + 1.5 * dist_IQR

dist_outliers = data[(data.distance_from_home <= lower_bound) | (data.distance_from_home >= upper_bound)]
dist_outliers
```

For the credit dataset perform

    Spot outliers in Income using bivariate plot
    Spot outliers in any one feature using box plot
    Detect outliers in any one features using IQR method
    Treat them using imputation (mean,median,mode)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\credit_risk_dataset.csv").head(20)


# plt.figure(figsize=(10,7))
plt.scatter(x = data['person_age'],y = data['person_income'], color = "k",label='income')
plt.xlabel('Age of the persons')
plt.ylabel("Income of the persons")
plt.title("Bivariate plot for age and income")
plt.legend()
plt.show()


# plt.figure(figsize=(20,15))
sns.boxplot(data['person_income'])
plt.xlabel("Person income")
plt.ylabel("Values")
plt.title("Box plot to detect outliers in the salary dataset")
plt.show()


#IQR
q1 = data['person_income'].quantile(.25)
q3 = data['person_income'].quantile(.75)

IQR = q3-q1

lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR

outliers = data[(data['person_income'] >= upper_bound) | (data['person_income'] <= lower_bound)]
print("\n\nOutliers using IQR method is \n",outliers)


# print("\n\nNumber of outliers in the person income column is \n",outliers['person_income'].count())    
# # mean_outlier = data['person_income'][(data['person_income'] >= upper_bound) | (data['person_income'] <= lower_bound)]
# data.loc[outliers,'person_income'] = data['person_income'].mean()
mean = data['person_income'].mean()
for i in range(len(data)):
    if (data['person_income'][i][(data['person_income'][i] >= upper_bound) | (data['person_income'][i] <= lower_bound)]):
        data['person_income'][i] = mean
# print(data)
sns.boxplot(data['person_income'])
plt.title("After imputing with mean")
plt.show()


data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\credit_risk_dataset.csv").head(20)

# print("\n\nOutliers\n",outliers['person_income'].count())    
median = data['person_income'].median()
for i in range(len(data)):
    if (data['person_income'][i][(data['person_income'][i] >= upper_bound) | (data['person_income'][i] <= lower_bound)]):
        data['person_income'][i] = median
# print(data)
sns.boxplot(data['person_income'])
plt.title("After imputing with median")
plt.show()


data = pd.read_csv("C:\\Users\\Admin\\Desktop\\Folders\\Datasets\\credit_risk_dataset.csv").head(20)

# print("\n\nOutliers\n",outliers['person_income'].count())    
zero = 0
for i in range(len(data)):
    if (data['person_income'][i][(data['person_income'][i] >= upper_bound) | (data['person_income'][i] <= lower_bound)]):
        data['person_income'][i] = zero
# print(data)
sns.boxplot(data['person_income'])
plt.title("After imputing with Zero(0)")
plt.show()
```
